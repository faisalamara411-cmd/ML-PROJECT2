```python
#train test split
from sklearn.model_selection import train_test_split
x=df.drop('ckd',axis=1)
y=df['ckd']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#list the numerical columns
num_cols=['age','blood_pressure','specific_gravity','blood_glucose_random','blood_urea','serum_creatinine','sodium','potassium','hemoglobin','packed_cell_volume','white_blood_cell_count','red_blood_cell_count']
```

```python
#satandard scaler
from sklearn.preprocessing import StandardScaler

# copy the original train and test test
x_train_scaled=x_train.copy()
x_test_scaled=x_test.copy()

# apply transformation
scaler=StandardScaler()

x_train_scaled[num_cols]=scaler.fit_transform(x_train[num_cols])
x_test_scaled[num_cols]=scaler.transform(x_test[num_cols])
```

Hyper parameter tunning

```python
# hyper parameter tuning using grid search cv and random grid search cv
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
from scipy.stats import uniform, randint

log_reg_params = {
    "C": [0.01, 0.1, 1, 10, 100],
    "penalty": ["l1", "l2"],
    "solver": ["liblinear", "saga"]
}

log_reg_grid = GridSearchCV(log_reg, log_reg_params, cv=5, scoring="accuracy", n_jobs=-1)
log_reg_grid.fit(x_train_scaled, y_train)

log_reg_random = RandomizedSearchCV(
    log_reg,
    {"C": uniform(0.01, 100), "penalty": ["l1", "l2"], "solver": ["liblinear", "saga"]},
    n_iter=20, cv=5, scoring="accuracy", random_state=42, n_jobs=-1
)
log_reg_random.fit(x_train_scaled, y_train)
```

```python
from scipy.stats import randint

dt = DecisionTreeClassifier(random_state=42)

dt_params = {
    "max_depth": [None, 5, 10, 20],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "criterion": ["gini", "entropy"]
}

dt_grid = GridSearchCV(dt, dt_params, cv=5, scoring="accuracy", n_jobs=-1)
dt_grid.fit(x_train, y_train)

dt_random = RandomizedSearchCV(
    dt,
    {"max_depth": randint(3, 20),
     "min_samples_split": randint(2, 20),
     "min_samples_leaf": randint(1, 10),
     "criterion": ["gini", "entropy"]},
    n_iter=20, cv=5, scoring="accuracy", random_state=42, n_jobs=-1
)
dt_random.fit(x_train, y_train)
```

```python
from scipy.stats import randint

rf = RandomForestClassifier(random_state=42)

rf_params = {
    "n_estimators": [100, 200, 500],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"]
}

rf_grid = GridSearchCV(rf, rf_params, cv=5, scoring="accuracy", n_jobs=-1)
rf_grid.fit(x_train, y_train)

rf_random = RandomizedSearchCV(
    rf,
    {"n_estimators": randint(100, 1000),
     "max_depth": randint(5, 50),
     "min_samples_split": randint(2, 20),
     "min_samples_leaf": randint(1, 10),
     "max_features": ["sqrt", "log2", None]},
    n_iter=20, cv=5, scoring="accuracy", random_state=42, n_jobs=-1
)
rf_random.fit(x_train, y_train)
```

```python
knn = KNeighborsClassifier()

knn_params = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan", "minkowski"]
}

knn_grid = GridSearchCV(knn, knn_params, cv=5, scoring="accuracy", n_jobs=-1)
knn_grid.fit(x_train_scaled, y_train)

knn_random = RandomizedSearchCV(
    knn,
    {"n_neighbors": randint(3, 30),
     "weights": ["uniform", "distance"],
     "metric": ["euclidean", "manhattan", "minkowski"]},
    n_iter=20, cv=5, scoring="accuracy", random_state=42, n_jobs=-1
)
knn_random.fit(x_train_scaled, y_train)
```

Model summary

```python

print("Logistic Regression Grid Best:", log_reg_grid.best_params_, log_reg_grid.best_score_)
print("Logistic Regression Random Best:", log_reg_random.best_params_, log_reg_random.best_score_)

print("Decision Tree Grid Best:", dt_grid.best_params_, dt_grid.best_score_)
print("Decision Tree Random Best:", dt_random.best_params_, dt_random.best_score_)

print("Random Forest Grid Best:", rf_grid.best_params_, rf_grid.best_score_)
print("Random Forest Random Best:", rf_random.best_params_, rf_random.best_score_)

print("KNN Grid Best:", knn_grid.best_params_, knn_grid.best_score_)
print("KNN Random Best:", knn_random.best_params_, knn_random.best_score_)
```

Interpretation:

Random forest performs almost the same as Decision Tree, but with more stability and less overfitting risk.


Best model

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

best_log_reg = log_reg_grid.best_estimator_
best_dt = dt_grid.best_estimator_
best_rf = rf_grid.best_estimator_
best_knn = knn_grid.best_estimator_

models = {
    "Logistic Regression": (best_log_reg, x_test_scaled),
    "Decision Tree": (best_dt, x_test),
    "Random Forest": (best_rf, x_test),
    "KNN": (best_knn, x_test_scaled)
}

for name, (model, Xdata) in models.items():
    y_pred = model.predict(Xdata)
    print(f"\n{name} Test Results:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Random forest has highest accuracy.
